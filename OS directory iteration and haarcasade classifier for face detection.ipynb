```python
import cv2
import os 
import matplotlib.pyplot as plt
```

```python
# Code for extracting faces from images using pretrained Haarcascade
i = 0
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
folder = "Test Images"

# Read the input image
# Resize
for image in os.listdir(folder):
       if image is not None:
            img = cv2.imread(os.path.join(folder,image))
            img = cv2.resize(img, [250, 250])  
            gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            
            plt.imshow(img[:,:,::-1])
            plt.xlabel(f'original image:{image}')
            plt.show()
            
            face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')      
            faces = face_cascade.detectMultiScale(gray, 1.1, 3)
            for (x, y, w, h) in faces:
                cv2.rectangle(img, (x, y), (x+w, y+h), (0, 0, 100), 0)
                faces = img[y:y + h, x:x + w]
                cv2.imwrite('ResizedFaces/face%i.jpg' %i, faces)
               
                plt.imshow(faces[:,:,::-1])
                plt.xlabel(f'face extracted:{image}')
                plt.show()
                #cv2.imwrite('ResizedFaces/detcted%i.jpg', img) # THIS LINE IS USED TO SAVE IMAGE TO FILE
                i += 1
                #plt.imshow('img', img)
                cv2.waitKey(10)
```
